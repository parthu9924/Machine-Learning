```python
class LinearRegression:
    
    def __init__(self,x,y):
        self.x=x
        self.y=y
        self.m=0
        self.c=0
    
    def predict(self,x=[]):
        
        if not x:
            x=self.x;
        y_pred=np.array([])
        for x_val in x:
            y_pred=np.append(y_pred,self.c+(self.m*x_val))
        
        return y_pred
    
    def cost_fucntion(self,y_pred,m,c):
        l=len(self.y)
        y=self.y
        j=(1/(2*l)*(np.sum(y_pred-y)**2))
        return j
    
    def update_coefficient(self,learning_rate):
        l=len(self.y)
        y=self.y;
        y_pred1=self.predict()
        y_pred2=self.predict();
        temp=(1/l)*(np.sum((y_pred1-y)*self.x))
        self.m=self.m-(learning_rate*temp)
        temp=(1/l)*(np.sum(y_pred2-y))
        self.c=self.c-(learning_rate*temp)
        
    def plot_linearRegression(self,y_pred):
        
        plt.scatter(self.x,self.y,color="red",label="accurate point")
        plt.plot(self.x,y_pred,color="blue",label="precicted line")
        plt.title("LinearRegression m="+str(self.m)+" and c = "+str(self.c))
        plt.xlabel("Area")
        plt.ylabel("price")
        plt.legend()
        plt.show()
    
    def get_accuracy(self,y_pred):
        
        a=self.y;
        e=y_pred;
        l=len(y_pred)
        return 1-sum([abs(a[i]-e[i])/e[i] for i in range(l) if e[i]!=0])/l
```

```python
def main():
    X = np.array([i for i in range(11)]) 
    Y = np.array([2*i for i in range(11)])
    linearRegression=LinearRegression(X,Y)
    iteration=10
    learning_rate=0.01
    
    i=0
    print("intial graph : ")
    y_pred=linearRegression.predict()
    linearRegression.plot_linearRegression(y_pred)
    print("initial accuracy : "+str(linearRegression.get_accuracy(y_pred)))
    
    while i<=iteration:
        y_pred=linearRegression.predict()
        #linearRegression.plot_linearRegression(y_pred)
        linearRegression.update_coefficient(learning_rate)
        i=i+1;
    
    print("final graph : ")
    y_pred=linearRegression.predict()
    linearRegression.plot_linearRegression(y_pred)
    print("final accuracy : "+str(linearRegression.get_accuracy(y_pred)))
    
    
if __name__=='__main__':
    main()
    
```
